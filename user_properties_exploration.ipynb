```python
from chesscom.user_properties.v1.user_properties_service_pb2 import (
    SearchUserPropertiesRequest,
    UserPropertiesPropertiesData,
)
from chesscom.user_targeting.v1.user_targeting_pb2 import (
    BinaryOperator,
    Criterion,
    LogicalOperator,
)

import chip
```

```python
# Load api keys
from dotenv import load_dotenv

load_dotenv()
```

```python
request = SearchUserPropertiesRequest(
    criteria=[
        Criterion(
            user_property="username",
            binary_operator=BinaryOperator.EQUALS,
            value="MangosCarloson",
        ),
    ],
    # properties = ['id'],
    logical_operator=LogicalOperator.AND,
    limit=5000,
)
```

```python
cq = chip.data.agi_funcs.ChessQuery()
```

```python
response = cq.fetch_up_page(request=request)
response
```

```python
user = list(response.user_properties)[0]
user
```

```python
import json

with open("player_properties.txt", "w") as f:
    json.dump(user, f, indent=4, sort_keys=True)
```
